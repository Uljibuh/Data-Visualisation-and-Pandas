```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import numpy as np
import pandas as pd

iris_df = pd.read_csv("/kaggle/input/iris/Iris.csv")
gap_df = pd.read_csv("../input/gapminder-datacamp-2007/gapminder_full.csv")

```

```python
# basic scatter plot
fig = px.scatter(iris_df, x="SepalWidthCm", y="SepalLengthCm")
fig.show()
```

```python
# Scatter plots with variables-size circular markers

fig = px.scatter(data_frame=iris_df, 
                 x="SepalWidthCm",
                 y="SepalLengthCm",
                    
                 color="Species",
                 size="PetalLengthCm",
                 hover_data=['PetalWidthCm'])

fig.show()
```

```python
# continous color scatter plots

fig = px.scatter(data_frame=iris_df,
                x="SepalWidthCm",
                y="SepalLengthCm",
                size="PetalLengthCm",
                color='PetalLengthCm')
fig.show()
```

```python
# scatter plots with symbles
fig = px.scatter(data_frame=iris_df,
                x="SepalWidthCm",
                y="SepalLengthCm",
                color='Species',
                symbol="Species")
fig.show()
```

```python
# marginal Distribution scatter plots
fig = px.scatter(data_frame=iris_df,
                x="SepalWidthCm",
                y="SepalLengthCm",
                size="PetalLengthCm",
                color='PetalLengthCm',
                marginal_x="histogram",
                marginal_y="box")
fig.show()
```

```python
px.scatter(data_frame=gap_df,
           x="gdp_cap",
           y="life_exp", animation_frame="year",
           animation_group="country",
           size="population",
           color="continent",
           hover_name="country",
           log_x=True, size_max=40, range_x=[100,100000], range_y=[25,90] )
```

```python
#basic line plot
fig = px.line(data_frame=gap_df.query("country=='Norway'"),
              x="year",
              y="life_exp",
              color='country')
fig.show()
```

```python
fig = px.line(data_frame=gap_df.query("continent=='Asia'"),
              x="year",
              y="life_exp",
              color='country')
fig.show()
```

```python
# line plot with markers (symbols)

fig = px.line(data_frame= gap_df.query("continent=='Oceania'"),
              x="year", 
              y="life_exp", 
              markers =True,
              #symbol = "country",
              color='country')
fig.show()
```

```python
fig = px.pie(data_frame= gap_df.query("year == 2007").query("continent == 'Europe'"), 
             values='population', 
             names='country',
             title='Population of Europe continent',
             hover_data=['life_exp'], 
             labels={'lifeExp':'life expectancy'})
fig.update_traces(textposition='inside', 
                  textinfo='percent+label')
fig.show()
```

```python
fig = px.bar(data_frame= gap_df.query("country == 'Sweden'"),
             x='year',
             y='population')
fig.show()
```

```python
fig = px.bar(data_frame= gap_df.query("country == 'Sweden'"),
             x='year', 
             y='population',
             hover_data=['life_exp', 'gdp_cap'], 
             )
fig.update_traces(marker_color ='rgb(158,202,225)', marker_line_color='rgb(8,48,107)',
                  marker_line_width=1.5, opacity=0.6)
fig.show()
```

```python
#Bar Plot with continous Color scale

fig = px.bar(gap_df.query("country=='Poland'"),
            x='year',
            y='population',
            hover_data=['life_exp', 'gdp_cap'],
            color='life_exp',
            color_continuous_scale="Emrld")
fig.show()
```

```python
# horizontal bar plot
fig = px.bar(gap_df.query("country=='China'"),
            x='year',
            y='population',
            hover_data=['life_exp', 'gdp_cap'],
            color='life_exp',
            color_continuous_scale="Emrld", orientation="h")
fig.show()
```

```python
N = 30
c = ['hsl('+str(h)+',50%'+',50%)' for h in np.linspace(0, 360, N)]

fig = go.Figure(
    data=[go.Box(y=3.5 * np.sin(np.pi * i/N) + i/N + (1.5 + 0.5 * np.cos(np.pi*i/N)) * np.random.rand(10),marker_color=c[i]) for i in range(int(N))])

fig.update_layout(
    xaxis=dict(showgrid=False, zeroline=False, showticklabels=False),
    yaxis=dict(zeroline=False, gridcolor='white'),
    paper_bgcolor='rgb(255,255,255)',
    plot_bgcolor='rgb(255,255,255)',
)

fig.show()
```

```python
# stacked bar plot
x= [1,3,5,7,9,11,13,47,17] 
y1 = [21,11,17,32,57,12]
y2= [17,18,29,40,45,23]

fig = go.Figure(data=[
    go.Bar(x=x, y=y1),
    go.Bar(x=x, y=y2)
])
fig.update_layout(barmode='stack')
fig.show()
```

```python
# grouped bar plot
x= [1,3,5,7,9,11,13,47,17] 
y1 = [21,11,17,32,57,12]
y2= [17,18,29,40,45,23]

fig = go.Figure(data=[
    go.Bar(x=x, y=y1),
    go.Bar(x=x, y=y2)
])
fig.update_layout(barmode='group')
fig.show()
```
