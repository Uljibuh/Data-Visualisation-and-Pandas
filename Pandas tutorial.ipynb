```python
import pandas as pd
import numpy as np
```

```python
people = {
    'first': ['Corey', 'Jane', 'John', 'Chris', np.nan, None, 'NA'], 
    'last': ['Schafer', 'Doe', 'Doe', 'Schafer', np.nan, np.nan, 'Missing'], 
    'email': ['CoreyMSchafer@gmail.com', 'JaneDoe@email.com', 'JohnDoe@email.com', None, np.nan, 'Anonymous@email.com', 'NA'],
    'age': ['33', '55', '63', '36', None, None, 'Missing']
}
```

```python
df = pd.DataFrame(people)
```

```python
df
```

```python
df.replace('NA', np.nan, inplace=True)
df.replace('Missing', np.nan, inplace=True)
```

```python
df
```

```python
df.dropna()
```

```python
df.dropna(axis='columns', how='all')
#df.dropna(axis='rows', how='any')
```

```python
df.dropna(axis='index', how='any', subset=['email'])
# if we need email address even we do not have names,
# make 'subset' argumenrt in code
```

```python
# whether values classifieda as nan
df.isna()
```

```python
# fill nan value with 0
df.fillna(0)
```

```python
df.dtypes
```

```python
# because 'age' in df is a object(string or mixed)
# this code will give you error
df['age'].mean()
```

```python
# we need to convert string to float 
df['age'] = df['age'].astype(float)
```
